```python
import pandas as pd
from sklearn.model_selection import train_test_split
from pathlib import Path

# Config
input_file = Path("../data/processed/sentences_cleaned.jsonl") # Point this to your file
output_dir = Path("../data/splits")
output_dir.mkdir(parents=True, exist_ok=True)

# 1. Load Data
print(f"Loading data from {input_file}...")
df = pd.read_json(input_file, lines=True)

# 2. Split Data (80% Train, 10% Val, 10% Test)
# First split: Separate out the Test set (10%)
train_val, test = train_test_split(df, test_size=0.10, random_state=42, shuffle=True)

# Second split: Separate Train and Validation (from the remaining 90%)
# 0.111 of 90% is approx 10% of the total
train, val = train_test_split(train_val, test_size=0.1111, random_state=42, shuffle=True) 

# 3. Save Splits
print(f"Saving splits to {output_dir}...")
train.to_json(output_dir / "train.jsonl", orient="records", lines=True, force_ascii=False)
val.to_json(output_dir / "val.jsonl", orient="records", lines=True, force_ascii=False)
test.to_json(output_dir / "test.jsonl", orient="records", lines=True, force_ascii=False)

print(f"Done! Stats:\nTrain: {len(train)}\nVal:   {len(val)}\nTest:  {len(test)}")
```
